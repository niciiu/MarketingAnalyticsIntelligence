```python
# Environment & Paths (robust to notebook location)

from pathlib import Path
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["axes.grid"] = True
```

```python
# Detect project root
CWD = Path.cwd()
ROOT = CWD if (CWD / "data_raw").exists() else CWD.parent

RAW_CSV   = ROOT / "data_raw" / "marketing_campaign_dataset.csv"   # INPUT (mentah)
CLEAN_CSV = ROOT / "data"     / "marketing_campaign_performance.csv"  # OUTPUT (clean)
ROI_CSV   = ROOT / "data"     / "roi_stability_summary.csv"           # OUTPUT (opsional)
DB_PATH   = ROOT / "data"     / "ads_analytics.db"                    # Dibuat di ETL terpisah

os.makedirs(ROOT / "data_raw", exist_ok=True)
os.makedirs(ROOT / "data", exist_ok=True)
os.makedirs(ROOT / "reports", exist_ok=True)

print("Project ROOT :", ROOT)
print("Raw exists   :", RAW_CSV.exists())
print("RAW_CSV path :", RAW_CSV)
```

# Load Data & First Look
- Tujuan: memahami struktur, tipe data, dan ukuran dataset untuk memandu cleaning


```python
df = pd.read_csv(RAW_CSV)
print("Shape RAW:", df.shape)
display(df.head(5))
display(df.describe(include="all").T.head(20))
df.info()
```

# Business-Grade Cleaning & Type Casting

- Deterministic & repeatable : aturan eksplisit, siap di-replikasi.
- Defensive casting : ubah teks uang menjadi float, tanggal menjadi datetime.
- Integrity checks : tidak ada nilai negatif untuk metrik biaya/volume.
- Idempotent : aman dijalankan ulang tanpa efek samping buruk.

```python
# Validasi kolom minimal yang dibutuhkan untuk KPI & EDA
required = [
    "Campaign_ID","Company","Campaign_Type","Target_Audience","Duration",
    "Channel_Used","Conversion_Rate","Acquisition_Cost","ROI","Location",
    "Language","Clicks","Impressions","Engagement_Score","Customer_Segment","Date"
]
missing = [c for c in required if c not in df.columns]
if missing:
    raise ValueError(f"Kolom wajib hilang: {missing}")
```

```python
# Normalisasi nilai uang — hilangkan simbol $ dan koma, jadikan float
df["Acquisition_Cost"] = (
    df["Acquisition_Cost"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
#Tanggal to datetime
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

```python
# Nilai numerik anti-negatif
for col in ["Acquisition_Cost","Clicks","Impressions","Engagement_Score","Conversion_Rate","ROI"]:
    if col in df.columns:
        df.loc[df[col] < 0, col] = 0
```

```python
# Drop baris invalid & duplikat exact
before = len(df)
df = df[df["Date"].notna()]
df = df.drop_duplicates()
removed = before - len(df)
```

```python
print(f"Records removed (invalid/dupe): {removed}")
print("Shape CLEAN (pre-derive):", df.shape)
```

```python
# Standarisasi nama kolom (snake case untuk downstream SQL/Streamlit)
df = df.rename(columns={
    "Campaign_ID": "campaign_id",
    "Company": "company",
    "Campaign_Type": "campaign_type",
    "Target_Audience": "target_audience",
    "Duration": "duration",
    "Channel_Used": "channel",
    "Conversion_Rate": "conversion_rate",
    "Acquisition_Cost": "spend",
    "ROI": "roi",
    "Location": "location",
    "Language": "language",
    "Clicks": "clicks",
    "Impressions": "impressions",
    "Engagement_Score": "engagement_score",
    "Customer_Segment": "customer_segment",
    "Date": "date",
})

display(df.head(3))
```

# ##KPI Derivation (Defensive)
- Definisi (turunan):
- CTR = clicks / impressions  
- CPC = spend / clicks  
- CPM = spend / impressions * 1000  
- Engagement Rate = engagement_score / impressions  
- ROAS = revenue / spend *(tidak tersedia di dataset ini maka skip)*  
- ROI = sudah disediakan pada dataset (akan dipakai sebagai metrik outcome)  

_apabila ada revenue di masa depan, tinggal tambahkan kolom itu & turunkan ROAS/ROI versi revenue_


```python
# Pastikan tipe numerik aman
for c in ["spend","clicks","impressions","engagement_score","conversion_rate","roi"]:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)
```

```python
# Turunan KPI defensif
df["ctr"]  = np.where(df["impressions"] > 0, df["clicks"] / df["impressions"], 0.0)
df["cpc"]  = np.where(df["clicks"]      > 0, df["spend"]  / df["clicks"],      0.0)
df["cpm"]  = np.where(df["impressions"] > 0, df["spend"]  / df["impressions"] * 1000.0, 0.0)
df["engagement_rate"] = np.where(df["impressions"] > 0, df["engagement_score"] / df["impressions"], 0.0)
```

```python
#Simpan CLEAN untuk downstream
df = df.sort_values("date")
CLEAN_CSV.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(CLEAN_CSV, index=False)
print(f"Clean dataset saved to {CLEAN_CSV}")
display(df.head(5))
```


# EDA — Executive View (ringkas & tajam)
goal pertanyaan: 
- Bagaimana distribusi ROI dan metrik efisiensi (CPC/CPM)?
- Apakah ROI cenderung naik seiring skala (Spend/Impressions)?
- Siapa channel/campaign yang konsisten kuat berdasarkan median/varians?


```python
#Distribusi & ringkasan numerik
fig, ax = plt.subplots()
ax.hist(df["roi"], bins=50)
ax.set_title("Distribution of ROI"); ax.set_xlabel("ROI"); ax.set_ylabel("Count")
plt.show()
```

```python
summary = df[["spend","impressions","clicks","ctr","cpc","cpm","roi","engagement_rate"]].describe().T
print("Numerical Summary (selected):")
display(summary)
```

```python
#Tren waktu — aggr harian spend vs ROI rata-rata
ts = df.groupby("date").agg(
    spend=("spend","sum"),
    roi_mean=("roi","mean")
).reset_index()

fig, ax1 = plt.subplots()
ax1.plot(ts["date"], ts["spend"])
ax1.set_ylabel("Spend (sum)")
ax1.set_title("Daily Spend (sum) & ROI (mean)")
ax2 = ax1.twinx()
ax2.plot(ts["date"], ts["roi_mean"])
ax2.set_ylabel("ROI (mean)")
plt.show()
```

```python
#Top channel — median ROI & variabilitas
chan = df.groupby("channel").agg(
    spend=("spend","sum"),
    roi_median=("roi","median"),
    roi_std=("roi","std"),
    n=("roi","count")
).reset_index().sort_values(["roi_median","spend"], ascending=[False, False])

print("Channel ranking (median ROI first):")
display(chan.head(10))
```

**Visual ROI Diagnostics — Insight**
Tujuan: memetakan efisiensi skala dan stabilitas efektifitas.
- **ROI vs Spend** → apakah skala memperbaiki atau menurunkan ROI?
- **ROI by Channel (boxplot)** → stabilitas antar channel
- **ROI Stability per Channel × Month** → bahan *risk view* (CV & MoM)

```python
#ROI vs Spend (scatter) ---
plt.figure(figsize=(10,5))
plt.scatter(df["spend"], df["roi"], alpha=0.25)
plt.title("ROI vs Spend — Scale Efficiency")
plt.xlabel("Spend"); plt.ylabel("ROI")
plt.tight_layout()
plt.show()
```

```python
#ROI by Channel (boxplot sederhana)
channels = sorted(df["channel"].dropna().unique())
bp_data  = [df.loc[df["channel"]==c, "roi"].values for c in channels]

plt.figure(figsize=(10,5))
plt.boxplot(bp_data, labels=channels, vert=True, showfliers=False)
plt.title("ROI Distribution by Channel")
plt.ylabel("ROI"); plt.xticks(rotation=25)
plt.tight_layout()
plt.show()
```

```python

# Stability per Channel × Month (CV & MoM) 
tmp = df.copy()
tmp["month_id"] = pd.to_datetime(tmp["date"]).dt.to_period("M").astype(str)

stability = tmp.groupby(["channel","month_id"]).agg(
    spend=("spend","sum"),
    impressions=("impressions","sum"),
    n=("roi","count"),
    roi_mean=("roi","mean"),
    roi_std=("roi","std")
).reset_index()

stability["roi_std"] = stability["roi_std"].fillna(0.0)
stability["roi_cv"]  = np.where(stability["roi_mean"]!=0, stability["roi_std"]/np.abs(stability["roi_mean"]), np.nan)
stability = stability.sort_values(["channel","month_id"])
stability["roi_mom"] = stability.groupby("channel")["roi_mean"].pct_change()

# Simpan untuk Streamlit / SQL diagnostics
Path(ROI_CSV).parent.mkdir(parents=True, exist_ok=True)
stability.to_csv(ROI_CSV, index=False)

print(f" ROI stability summary saved → {ROI_CSV}")
display(stability.head(10))

```

```python
#Visual: ROI Mean over Time per Channel ---
plt.figure(figsize=(10,5))
for ch in channels:
    sub = stability[stability["channel"]==ch]
    plt.plot(sub["month_id"], sub["roi_mean"], label=ch)
plt.title("ROI Mean by Channel Over Time")
plt.xlabel("Month"); plt.ylabel("ROI (mean)")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- 5.5 Executive quick-table: Top stable channels (low CV, with scale) ---
# filter: hanya channel dengan spend ≥ P50 spend
p50_spend = stability.groupby("channel")["spend"].sum().median()
stable_rank = (stability.groupby("channel")
               .agg(total_spend=("spend","sum"),
                    roi_mean=("roi_mean","mean"),
                    roi_cv=("roi_cv","mean"),
                    months=("month_id","nunique"))
               .reset_index())
stable_rank = stable_rank[stable_rank["total_spend"] >= p50_spend]
stable_rank = stable_rank.sort_values(["roi_cv","roi_mean"], ascending=[True, False])

print("Top Stable Channels (low CV, sufficient scale):")
display(stable_rank.head(10))
```

# Econometric Diagnostics — OLS (directional check)
_*Pertanyaan:* Seberapa besar kontribusi **scale inputs** (Spend, Impressions, Clicks) terhadap ROI?_
- Gunakan **OLS sederhana** untuk directional insight, bukan kausal.
_Hasil yang dicari:_
- Sign & signifikansi koefisien (apakah naik/turun ROI saat variabel naik)
- R-squared (kekuatan penjelasan model)
- Catatan: ROI bisa non-linear terhadap spend, gunakan juga transformasi log jika data mendukung.


```python
try:
    import statsmodels.api as sm
    import numpy as np

    # Model linear dasar
    X = df[["spend","impressions","clicks"]].copy()
    X = sm.add_constant(X)
    y = df["roi"]
    model_lin = sm.OLS(y, X, missing="drop").fit()
    print("=== OLS Linear: ROI ~ spend + impressions + clicks ===")
    print(model_lin.summary())
    
# Model log-log (jika semua > 0)
    if (df[["spend","impressions","clicks","roi"]] > 0).all().all():
        Xlog = np.log(df[["spend","impressions","clicks"]])
        Xlog = sm.add_constant(Xlog)
        ylog = np.log(df["roi"])
        model_log = sm.OLS(ylog, Xlog, missing="drop").fit()
        print("\n=== OLS Log-Log: log(ROI) ~ log(spend) + log(impressions) + log(clicks) ===")
        print(model_log.summary())
    else:
        print("\n(Log-log model dilewati karena ada nilai ≤ 0)")

except Exception as e:
    print(" Econometric diagnostics skipped:", e)
```

#  Executive Insight — 1-Minute Brief (auto-generated)
_**How to read:**_ bullet point di bawah menarik temuan inti (stable baseline, scale efficiency, dan channel priority).


```python
# Metrics (defensive)
roi_mean = float(df["roi"].mean())
roi_std  = float(df["roi"].std())
corr_spend_roi = float(df[["spend","roi"]].corr(numeric_only=True).iloc[0,1])

if "stable_rank" in locals() and not stable_rank.empty:
    top_ch = stable_rank.head(1)["channel"].iloc[0]
else:
    top_ch = df["channel"].mode().iloc[0] if "channel" in df.columns else "N/A"

# Klasifikasi arah korelasi dengan threshold kecil untuk mendekati nol
def corr_label_id(v: float) -> str:
    if abs(v) < 0.02: return "mendekati nol"
    return "positif" if v > 0 else "negatif"

def corr_label_en(v: float) -> str:
    if abs(v) < 0.02: return "near-zero"
    return "positive" if v > 0 else "negative"

corr_id = corr_label_id(corr_spend_roi)
corr_en = corr_label_en(corr_spend_roi)

```

```python
#Compose summaries 
summary_id = (
    f"ROI baseline stabil pada {roi_mean:.2f}±{roi_std:.2f} (200K records) "
    f"dengan korelasi belanja dengan ROI {corr_id} ({corr_spend_roi:.2f}); "
    f"channel {top_ch} menunjukkan kombinasi efisiensi dan stabilitas terbaik. "
    f"Rekomendasi: pertahankan alokasi inti pada channel tersebut dan fokuskan peningkatan pada konversi pasca klik."
)

summary_en = (
    f"Baseline ROI remains stable at {roi_mean:.2f}±{roi_std:.2f} (200K records) "
    f"with a {corr_en} spend to ROI correlation ({corr_spend_roi:.2f}); "
    f"{top_ch} demonstrates the strongest blend of efficiency and stability. "
    f"Recommendation: sustain core allocation on this channel and prioritize post-click conversion improvements."
)

# --- Print once (clean) ---
print("Summary (ID):")
print(summary_id, "\n")
print("Summary (EN):")
print(summary_en)
```
